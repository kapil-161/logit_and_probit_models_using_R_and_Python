```r
library(aod)
library(ggplot2)
library(dummies)
library(dplyr)
library(regclass)
```

```r
mydata <- read.csv("Meritlist.csv")
## view the first few rows of the data
head(mydata)
```

```r
df=select(mydata, -c(District,Merit.No))
```

```r
df=dummy.data.frame(df)
```

```r
df
```

```r
df=select(df, -c(GenderF,SchoolO,SchoolNull))
```

```r
df
```

```r
df$Rampur <- ifelse(df$Escore >79, 1, 0)
```

```r
df
```

```r
summary(df)
```

```r
print(sapply(df, sd))
```

```r
mylogit <- glm( Rampur~ Province+GenderM+SchoolG+SchoolP+SLC+NEB, data = df, family = binomial)
```

```r
confusion_matrix(mylogit)
```

```r
summary(mylogit)
```

```r
confint(mylogit)
```

```r
exp(cbind(OR = coef(mylogit)))
```

```r
df$predict=predict(mylogit, data=df,type ="response")
```

```r
df
```

```r
df$predicted <- ifelse(df$predict >0.3, 1, 0)
```

```r
df
```

```r
mean(df$predicted== df$Rampur)
```

```r
df %>% count(predicted)
```
